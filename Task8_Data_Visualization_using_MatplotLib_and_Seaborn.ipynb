##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:



image.png

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([10,20,30])
y = np.array([20,40,10])
a = np.array([10,20,30])
b = np.array([40,10,30])
plt.plot(x,y,label ='line1-width-3',color='blue',linewidth =3)
plt.plot(a,b,label = 'line2-width-5',color ='red',linewidth =5)
plt.xticks(np.arange(10, 31, 5))
plt.xlim(10, max(x))
plt.ylim(10, max(y))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two or more lines with diffrent widths and colors with suitable legends')
plt.legend()
plt.show()
````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(languages,popularity,color="blue")
plt.ylim(bottom = 0, top=25)
plt.xlim(left = 0)
plt.title('Popularity of Programming Language\nWorldwide, Oct 2017 compared to a year ago')
plt.xlabel('Langages')
plt.ylabel('Popularity')
plt.grid(axis='y', linestyle='-', alpha=0.7 ,color='r')
plt.grid(axis='x', linestyle='--')
plt.minorticks_on()
plt.grid(which='minor', color='black', linestyle=':', linewidth=0.5)
plt.show()
````

##7.  Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.barh(languages, popularity, color='green',align='edge')

plt.xlim(0,25)
plt.ylim(0)

plt.xlabel('Popularity')
plt.ylabel('Languages')
plt.title('Popularity of Programming Languages\nWorldwide, Oct 2017 compared to a year ago')

plt.grid(True, which= 'major',color='red',linewidth=0.5)

plt.minorticks_on()
plt.grid(which='minor', color='black', linestyle=':', linewidth=0.5)

plt.tight_layout()

plt.show()
````

##Exercise 1: Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties: 
```

Y label name = Total profit
```

The line plot graph should look like this.



image.png

````python
import matplotlib.pyplot as plt
months = df['month_number']
total_profit = df['total_profit']
plt.figure(figsize=(10,6))
plt.plot(months,total_profit)
font1 = {'family':'serif','size':15}
font2 = {'family':'serif','size':20}
plt.xlabel('Month Number',fontdict = font1)
plt.ylabel('Total Profit',fontdict = font1)
plt.title('Company profit per month',fontdict = font2)

plt.xticks(ticks=months)
plt.show()
````

##Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

image.png

````python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("company_sales_data.csv")
month=df['month_number']
facecream=df['facecream']
facewash=df['facewash']
facecream_pos=[i-.25 for i in list(month)]
facewash_pos=[i for i in list(month)]

plt.figure(figsize=[10,6])
plt.bar(facecream_pos,facecream,width=.25,align='edge',label='Face cream sales data')
plt.bar(facewash_pos,facewash,width=.25,align='edge',label='Face wash sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
plt.title('Facewash and facecream series data')
plt.xticks(month)
plt.grid(linewidth=1,linestyle='--')
plt.legend(loc='upper left')

plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import matplotlib.pyplot as plt

total_sales = {
    'FaceCream'    : df['facecream'].sum(),
    'FaceWash'     : df['facewash'].sum(),
    'ToothPaste'   : df['toothpaste'].sum(),
    'Bathing soap' : df['bathingsoap'].sum(),
    'Shampoo'      : df['shampoo'].sum(),
    'Moisturizer'  : df['moisturizer'].sum()
}

plt.figure(figsize=(10,6))
plt.pie(total_sales.values(), labels = total_sales.keys(), autopct='%1.1f%%')
plt.legend(loc = 'lower right')
plt.title('Sales Data')
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
months             = df['month_number']
facecream_sales    = df['facecream']
facewash_sales     = df['facewash']
toothpaste_sales   = df['toothpaste']
bathingsoap_sales  = df['bathingsoap']
shampoo_sales      = df['shampoo']
moisturizer_sales  = df['moisturizer']

labels = ['face Cream','Face Wash','Tooth paste','Bathing soap','Shampoo','Moisturizer']
colors=['m','c','r','k','g','y']
plt.stackplot(months,facecream_sales,facewash_sales,toothpaste_sales,bathingsoap_sales,shampoo_sales,moisturizer_sales, labels=labels,colors =colors)
plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
plt.title('All products sales data using slack plot')
plt.legend()
plt.show()
````
